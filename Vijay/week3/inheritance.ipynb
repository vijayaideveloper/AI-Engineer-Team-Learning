```python
class Kollywood:

    def Details(self):
        self.work = "Acting"
        self.place = 'TamilNadu'
```

```python
class Vijay(Kollywood):

    def __init__(self):
        self.name = "Vijay"
        self.age = 51
        self.gender = 'Male'
        self.skill = ['Acting', 'Dancing', 'Singing', 'Respective']

    def info(self):
        self.Details()
        print(f'Name : {self.name}')
        print(f'Age : {self.age}')
        print(f'Gender : {self.gender}')
        print(f'Skills : {self.skill}')
    
class Rajini(Kollywood):

    def __init__(self):
        self.name = "Rajini"
        self.age = 75
        self.gender = 'Male'
        self.skill = ['Acting', 'Style', 'Mass']

    def info(self):
        self.Details()
        print(f'Name : {self.name}')
        print(f'Age : {self.age}')
        print(f'Gender : {self.gender}')
        print(f'Skills : {self.skill}')
```

```python
k1 = Vijay()
k1.info()
```

```python
k2 = Rajini()
k2.info()
```
